```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
files = os.listdir("stability/")
#files = [f for f in files if "1stn" in f]
print(files)

x = pd.DataFrame()
y = pd.DataFrame()

for i in [f for f in files if "compare" not in f]:
    x = pd.concat((x, pd.read_csv("stability/" + i[:-4] + "-compare.csv", index_col=0).sort_index()))
    y = pd.concat((y, pd.read_csv("stability/" + i, index_col=0).sort_index()))
    
x *= 4.18
```

```python
def fitfunc(z, a, b, c, d):
    return a*(z["SOLV"] + z["COUL"]) + b*z["LJ (SR)"] + c*z["SAS"] + d*z["-TS"]

def test_fit(z, a, b, c, d):
    return a*(z["SOLV"]+z["COUL"]) + b*z["LJ"] + c*z["SAS"] + d*z["-TS"]
```

```python
p = curve_fit(fitfunc, y, x["EXP"])[0]
print(p)

q = curve_fit(test_fit, x, x["EXP"])[0]
print(q)
```

```python
fit = y.loc[:, ["SOLV", "COUL", "LJ (SR)", "SAS", "-TS"]].copy()

fit.iloc[:, :2] *= p[0]
fit.iloc[:, 2] *= p[1]
fit.iloc[:, 3] *= p[2]
fit.iloc[:, 4] *= p[3]
#for i in range(len(p)):
#    fit.iloc[:, i] *= p[i]

plt.plot(x["EXP"], fit.sum(axis=1), '.')
print(np.corrcoef(x["EXP"], fit.sum(axis=1)))
```
